# **1.9 Menangani Missing Value**

**Tujuan**

> Menangani nilai kosong pada dataset agar seluruh variabel siap digunakan dalam proses analisis.



```python
# Menghapus data yang tidak memiliki nama produk
df = df.dropna(subset=['product_name'])

# Mengisi nilai kosong pada data kategorikal
df['brands'] = df['brands'].fillna('Unknown')
df['categories'] = df['categories'].fillna('Unknown')
df['countries'] = df['countries'].fillna('Unknown')

# Mengisi nilai kosong pada data numerik dengan nilai median
kolom_numerik = [
    'additives_n',
    'energy_100g',
    'fat_100g',
    'saturated-fat_100g',
    'sugars_100g',
    'fiber_100g',
    'proteins_100g',
    'salt_100g'
]

for kolom in kolom_numerik:
    df[kolom] = df[kolom].fillna(df[kolom].median())

# Mengisi nilai kosong pada nutrition grade
df['nutrition_grade_fr'] = df['nutrition_grade_fr'].fillna('Unknown')
```

```python
df.isnull().sum()
```

**Penjelasan**

> Proses penanganan *missing value* dilakukan dengan menghapus data yang tidak memiliki nama produk, mengisi nilai kosong pada variabel kategorikal menggunakan nilai **"Unknown"**, serta mengisi nilai kosong pada variabel numerik menggunakan median. Metode median dipilih karena lebih stabil terhadap nilai ekstrem ***(outlier)***. Setelah proses tersebut dilakukan, seluruh variabel tidak lagi memiliki *missing value*, sehingga dataset siap digunakan untuk tahap analisis selanjutnya.
